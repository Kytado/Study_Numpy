```python
import urllib.request
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Đường dẫn của ảnh
url = 'https://th.bing.com/th/id/OIP.-kK-VYlMy_U69cvusBfF7wHaE7?pid=ImgDet&rs=1'

# Tải ảnh từ URL
urllib.request.urlretrieve(url, "image.jpg")

# Đọc ảnh
img = cv2.imread('image.jpg')

# Hiển thị ảnh
cv2_imshow(img)

# Lấy kích thước của ảnh
height, width = img.shape[:2]
print("Chiều rộng của ảnh:", width)
print("Chiều cao của ảnh:", height)

# Thay đổi kích thước ảnh
resized_image = cv2.resize(img, (int(width/2), int(height/2)))
cv2_imshow(resized_image)

# Lấy giá trị màu tại một điểm ảnh
pixel = img[100,100]
print("Giá trị màu tại điểm (100,100) là:", pixel)

# Chuyển màu về trắng đen
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)

cv2.destroyAllWindows()
```
