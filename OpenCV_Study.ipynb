```python
import urllib.request
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# ng dn ca nh
url = 'https://th.bing.com/th/id/OIP.-kK-VYlMy_U69cvusBfF7wHaE7?pid=ImgDet&rs=1'

# Ti nh t URL
urllib.request.urlretrieve(url, "image.jpg")

# c nh
img = cv2.imread('image.jpg')

# Hin th nh
cv2_imshow(img)

# Ly kch thc ca nh
height, width = img.shape[:2]
print("Chiu rng ca nh:", width)
print("Chiu cao ca nh:", height)

# Thay i kch thc nh
resized_image = cv2.resize(img, (int(width/2), int(height/2)))
cv2_imshow(resized_image)

# Ly gi tr mu ti mt im nh
pixel = img[100,100]
print("Gi tr mu ti im (100,100) l:", pixel)

# Chuyn mu v trng en
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)

cv2.destroyAllWindows()
```
